```python
df.query('URL.str.contains("btv1b102209721")')
```

```python
df['ARK'] = df.URL.str.extract('ark:(?P<ark>.*$)')
```

```python
df['image_URL'] = df.ARK.apply(
    lambda ark: f'https://openapi.bnf.fr/iiif/image/v3/ark:{ark}/f1/full/,512/0/default.jpg'  # height of 512 px
)
```

```python
df
```

```python
df['image_filename'] = 'tn_BnF_ark' + df.ARK.str.replace('/', '_') + '.jpg'
```

We'll use curl to fetch the images and avoid downloading images that we already have, so we can run the script multiple times in case of failures due to network issues, server overload, or rate limits.

```python
with open('fetch_posters.txt', 'wt') as f:
    for i, image in df.iterrows():
        # -s silent; -S show errors; only download if the output doesn't exist here yet
        print(
            f"[ ! -f images/{image.image_filename} ] && curl -s -S --retry 5 '{image.image_URL}' -o images/{image.image_filename}",
            file=f,
        )
```
